```python
df["Combined_Text"] = (
    df["Resume_Text"].fillna("").astype(str) + " " +
    df["Skills"].fillna("").astype(str) + " " +
    df["projects"].fillna("").astype(str) + " " +
    df["Education"].fillna("").astype(str) + " " +
    df["Experience"].fillna("").astype(str) + " " +
    df["Certificates"].fillna("").astype(str)
)
```

```python
import re

def clean_text(text):
    text = str(text).lower()
    text = re.sub(r"http\S+", "", text)
    text = re.sub(r"[^a-zA-Z ]", " ", text)
    text = re.sub(r"\s+", " ", text)
    return text

df["Combined_Text"] = df["Combined_Text"].apply(clean_text)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer(
    stop_words="english",
    lowercase=True,
    max_features=50000,
    ngram_range=(1,2),
    min_df=1,
    max_df=0.95,
    sublinear_tf=True
)

X = vectorizer.fit_transform(df["Combined_Text"])

print("Vectorization Completed")
print("Shape:", X.shape)
```

```python
X = vectorizer.fit_transform(df["Combined_Text"])
```

```python
y = df["Applied_Role"]

print("Features Shape:", X.shape)
print("Labels Shape:", y.shape)
```

```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()

y = label_encoder.fit_transform(y)

print("Total Classes:", len(label_encoder.classes_))
print(label_encoder.classes_)
```

```python
with open("../models/tfidf_vectorizer.pkl","wb") as file:
    pickle.dump(vectorizer,file)

print("Vectorizer Saved Successfully")
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    random_state=42,
    stratify=y
)

print("Train Shape:", X_train.shape)
print("Test Shape:", X_test.shape)
```
