```python
import pandas as pd
import matplotlib.pyplot as plt

print("Libraries Imported Successfully")
```

```python
df = pd.read_csv("processed_resume_dataset.csv")

print("Dataset Loaded Successfully")
```

```python
print("Rows :", df.shape[0])
print("Columns :", df.shape[1])
```

```python
print(df.columns.tolist())
```

```python
df.info()
```

```python
df.describe(include="all")
```

```python
df.isnull().sum()
```

```python
print(df["Applied_Role"].unique())
```

```python
role_counts = df["Applied_Role"].value_counts()

print(role_counts)
```

```python
plt.figure(figsize=(10,6))

role_counts.plot(kind="bar")

plt.title("Role Distribution")

plt.xlabel("Applied Role")

plt.ylabel("Number of Candidates")

plt.xticks(rotation=45)

plt.show()
```

```python
plt.figure(figsize=(8,8))

role_counts.plot(
    kind="pie",
    autopct="%1.1f%%"
)

plt.ylabel("")

plt.title("Role Distribution")

plt.show()
```

```python
education = df["Education"].value_counts()

print(education.head(10))
```

```python
plt.figure(figsize=(12,6))

education.head(10).plot(kind="bar")

plt.title("Top Education Qualifications")

plt.xlabel("Education")

plt.ylabel("Count")

plt.xticks(rotation=45)

plt.show()
```

```python
skills = (
    df["Skills"]
    .dropna()
    .str.split(",")
    .explode()
    .str.strip()
)

skill_counts = skills.value_counts()

print(skill_counts.head(20))
```

```python
plt.figure(figsize=(12,6))

skill_counts.head(20).plot(kind="bar")

plt.title("Top 20 Skills")

plt.xlabel("Skills")

plt.ylabel("Frequency")

plt.xticks(rotation=60)

plt.show()
```

```python
certificates = (
    df["Certificates"]
    .dropna()
    .str.split(",")
    .explode()
    .str.strip()
)

certificate_counts = certificates.value_counts()

print(certificate_counts.head(15))
```

```python
plt.figure(figsize=(12,6))

certificate_counts.head(15).plot(kind="bar")

plt.title("Top Certificates")

plt.xlabel("Certificate")

plt.ylabel("Count")

plt.xticks(rotation=45)

plt.show()
```

```python
df["Resume_Length"] = df["Resume_Text"].apply(len)

print(df["Resume_Length"].describe())
```

```python
plt.figure(figsize=(10,6))

plt.hist(df["Resume_Length"], bins=20)

plt.title("Resume Length Distribution")

plt.xlabel("Characters")

plt.ylabel("Count")

plt.show()
```

```python
experience = df["Experience"].value_counts()

print(experience.head(10))
```

```python
plt.figure(figsize=(12,6))

experience.head(10).plot(kind="bar")

plt.title("Experience Distribution")

plt.xlabel("Experience")

plt.ylabel("Candidates")

plt.xticks(rotation=45)

plt.show()
```

```python
numeric_df = df.select_dtypes(include=["number"])

print(numeric_df.corr())
```
